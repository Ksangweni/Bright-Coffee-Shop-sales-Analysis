```sql
--A. Exploring the Data
    -- 1. View all the columns and rows 
    -- 2. Perform Row Count (total_transactions) = 14 116 total transactions
    -- 3. Check the Date Range of the data = 1 Jan 2023 - 30 Jun 2023 (6 months' data).
    -- 4. On which locations do we serve our product (Active_Locations) = Lower Manhattan/Hell's Kitchen/Astoria
    -- 5. How many products do we offer = 29 product categoies / 80 sub categories
  
SELECT *
FROM workspace.bright_learn.coffee_shop_sales;

SELECT COUNT(transaction_id) AS TOTAL_TRANSACTIONS
FROM workspace.bright_learn.coffee_shop_sales;

SELECT MIN(transaction_date) AS DAY_ONE
  FROM workspace.bright_learn.coffee_shop_sales;

SELECT MAX(transaction_date) AS LAST_DAY
  FROM workspace.bright_learn.coffee_shop_sales;  

SELECT DISTINCT(store_location) AS Active_locations,
       store_id
  FROM workspace.bright_learn.coffee_shop_sales;

SELECT DISTINCT (product_type) AS product_category
  FROM workspace.bright_learn.coffee_shop_sales; 

SELECT DISTINCT (product_detail) AS sub_category
  FROM workspace.bright_learn.coffee_shop_sales;       



--B. SALES TRENDS ANALYSIS BY DAY,
 
SELECT 
-- Dates 
        transaction_date,
        Dayname(transaction_date) AS Day_name,
        Monthname(transaction_date) AS Month_name,
        DAYofmonth(transaction_date) AS Month_day,
        date_format(transaction_time, 'HH:mm:ss') AS Purchase_time,
 
    CASE  
         WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '00:00:00' AND '08:59:59' THEN '01.Breakfast'
         WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '09:00:00' AND '11:59:59' THEN '02.Brunch'
         WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '12:00:00' AND '14:59:59' THEN '03.Lunch'
         WHEN date_format(transaction_time, 'HH:mm:ss') >= '15:00:00' THEN '04.Late lunch'
    END AS Day_breakdown, 

    CASE 
          WHEN Day_name IN ('Sun' , 'Sat') THEN 'Weekend'
          ELSE 'Weekday'
    END AS Day_classification,  


-- COUNTS OF IDS
        COUNT(DISTINCT transaction_id) AS Number_of_sales,
        COUNT(DISTINCT product_id) AS Number_of_products, 
        COUNT(DISTINCT store_id) AS Number_of_stores,

-- REVENUE
        transaction_qty,
        unit_price, 
        SUM(transaction_qty*unit_price) AS Revenue_per_day,
     
      CASE 
           WHEN revenue_per_day <= 50 THEN '01. Low Spend'
           WHEN revenue_per_day BETWEEN 51 AND 100 THEN '02. Medium Spend'
           ELSE '03.High Spend'
      END AS Spend_brackets,

-- CATEGORICAL COLUMNS
        store_location,
        product_category,
        product_detail
FROM
       workspace.bright_learn.coffee_shop_sales
GROUP BY 
        transaction_date, 
        Day_name,
        store_location,
        product_category,
        transaction_qty,
        unit_price,
        product_detail,
        Month_name,
        Purchase_time,
        Day_breakdown,
        Month_day;
      
```
